```python
#1 Scenario: A digital locker requires users to enter a numerical passcode. 
#As part of a security feature, the system checks the sum of the digits of the passcode.
#Write logic to compute the sum of the digits of a given number.
```

```python
def SumOfDigits(num):
    total=0
    while(num>0):
        digit=num%10
        total+=digit
        num=num//10
    return total
```

```python
num=int(input("Enter your passcod:"))
result=SumOfDigits(num)
print("Sum of digits:", result)
```

```python
def ValidateTransaction(TransactionID):
    reversedID=str(TransactionID)[::-1]
    if str(TransactionID==reversedID):
        return "Transaction ID is valid"
    else:
         return "Transaction ID is invalid"
```

```python
#3 In a secure login system, certain features are enabled only for users with prime-numbered user IDs.
#Write logic to check if a given number is prime.
```

```python
number=int(input("Enter a number: "))
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        print(number ,"%", i, ' = ',number%i)
        if number % i == 0:
            print("Not prime")
            return False
    print("Is prime")
    return True
```

```python
user_id = 17
print(f"User ID: {user_id}, Is prime: {is_prime(user_id)}")
```

```python
#4 A scientist is working on permutations and needs to calculate the factorial of numbers frequently.
 #Write logic to find the factorial of a given number using recursion.
```

```python
def FactorialRecursive(num):
        if (num < 0):
            return "Factorial is not defined for negative numbers."
        elif (num == 0):
            return 1
        else:
            return num * FactorialRecursive(num - 1)
```

```python
num=7
result=FactorialRecursive(num)
print("The factorial of given number is", result)
```

```python
#5 A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
#Write logic to check whether a given number is an Armstrong number.
```

```python
def Armstrong(num):
    string=str(num)
    digits=len(string)
    total=0
    for digit in string:
        total=total + int(digit)**digits
    if (total==num):
        return True
    else:
        return False 
```

```python
number=int(input("Enter a number"))
if Armstrong(number):
    print(number, "is an Armstrong number.")
else:
    print(number, "is not an Armstrong number.")
```

```python
#9 A plagiarism detection tool compares words from different documents and checks if they are anagrams 
#(same characters but different order).
#Write logic to check whether two given strings are anagrams.
```

```python
def are_anagrams_sorted(str1, str2):
    s1_normalized = str1.lower()
    s2_normalized = str2.lower()
    
    if len(s1_normalized) != len(s2_normalized):
        print("Not an anagram")
    else:
        sorted_s1 = sorted(s1_normalized)
        sorted_s2 = sorted(s2_normalized)
    
    if sorted_s1 == sorted_s2: 
        print("Anagram")
        return True 
    else:
        print("Not an anagram")
        return False
```

```python
s1 = "A gentleman"
s2 = "Elegant man"
```

```python
are_anagrams_sorted(s1, s2)
```
